```python
#Train-Test Split
from sklearn.model_selection import train_test_split

X = df[['Daily Time Spent on Site', 'Age', 'Area Income','Daily Internet Usage', 'Male']]
y = df['Clicked on Ad']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=101)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
print('df=',df.columns)
print ('X_train=',X_train.columns)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

def print_score(clf, X_train, y_train, X_test, y_test, train=True):
    if train:
        pred = clf.predict(X_train)
        clf_report = pd.DataFrame(classification_report(y_train, pred, output_dict=True))
        print("Train Result:\n================================================")
        print(f"Accuracy Score: {accuracy_score(y_train, pred) * 100:.2f}%")
        print("_______________________________________________")
        print(f"CLASSIFICATION REPORT:\n{clf_report}")
        print("_______________________________________________")
        print(f"Confusion Matrix: \n {confusion_matrix(y_train, pred)}\n")
        
    elif train==False:
        pred = clf.predict(X_test)
        clf_report = pd.DataFrame(classification_report(y_test, pred, output_dict=True))
        print("Test Result:\n================================================")        
        print(f"Accuracy Score: {accuracy_score(y_test, pred) * 100:.2f}%")
        print("_______________________________________________")
        print(f"CLASSIFICATION REPORT:\n{clf_report}")
        print("_______________________________________________")
        print(f"Confusion Matrix: \n {confusion_matrix(y_test, pred)}\n")
        import matplotlib.pyplot as plt
        import seaborn as sns
        plt.figure(figsize=(4, 4))
        sns.heatmap(confusion_matrix(y_test, pred),  annot=True,xticklabels=['Neg','pos'], yticklabels=['Neg','Pos'],fmt="d")
        plt.title("Confusion matrix")
        plt.ylabel('True class')
        plt.xlabel('Predicted class')
        plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

solver_list = ['liblinear', 'lbfgs','newton-cg', 'sag', 'saga']
params = dict(solver=solver_list)
log_reg = LogisticRegression(C=1, n_jobs=-1, random_state=34)

clf = GridSearchCV(log_reg, params, cv=5)

clf.fit(X_train, y_train)
scores = clf.cv_results_['mean_test_score']

print("Without Scaling:")
for score, solver in zip(scores, solver_list):
    print(f"  {solver} {score:.3f}" )
    
```

```python
log_clf_lbfgs=LogisticRegression(solver='lbfgs')
log_clf_lbfgs.fit(X_train,y_train)
print_score(log_clf_lbfgs,X_train, y_train, X_test, y_test, train=True)
print_score(log_clf_lbfgs,X_train, y_train, X_test, y_test, train=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
y_predict=log_clf_lbfgs.predict(X_test)
acc1=accuracy_score(y_test, y_predict) * 100
conf_matrix = confusion_matrix(y_test, y_predict)
plt.figure(figsize=(4, 4))
sns.heatmap(conf_matrix,  annot=True,xticklabels=['Neg','pos'], yticklabels=['Neg','Pos'],fmt="d")
plt.title("Confusion matrix_Without Normalization\nSolver:lbfgs, Accuracy=%.2f"%acc1)
plt.ylabel('True class')
plt.xlabel('Predicted class')
plt.show()
```

```python
log_clf_lbfgs.coef_
```

```python
log_clf_liblinear=LogisticRegression(solver='liblinear')
log_clf_liblinear.fit(X_train,y_train)
print_score(log_clf_liblinear,X_train, y_train, X_test, y_test, train=True)
print_score(log_clf_liblinear,X_train, y_train, X_test, y_test, train=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
y_predict_lib=log_clf_liblinear.predict(X_test)
acc2=accuracy_score(y_test, y_predict_lib) * 100
conf_matrix = confusion_matrix(y_test, y_predict_lib)
plt.figure(figsize=(4, 4))
sns.heatmap(conf_matrix,  annot=True,xticklabels=['Neg','pos'], yticklabels=['Neg','Pos'],fmt="d")
plt.title("Confusion matrix_Without Normalization\nSolver:liblinear, Accuracy=%.2f"%acc2)
plt.ylabel('True class')
plt.xlabel('Predicted class')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, OrdinalEncoder
from sklearn.compose import make_column_transformer
#Scaling
num_columns = ['Daily Time Spent on Site', 'Age', 'Area Income', 'Daily Internet Usage', 'Male']
ct = make_column_transformer(
    (MinMaxScaler(), num_columns),
    (StandardScaler(), num_columns),
    remainder='passthrough'
)

X_train_norm = ct.fit_transform(X_train)
X_test_norm = ct.fit_transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

solver_list = ['liblinear', 'lbfgs','newton-cg', 'sag', 'saga']
params = dict(solver=solver_list)
log_reg = LogisticRegression(C=1, n_jobs=-1, random_state=34)

clf = GridSearchCV(log_reg, params, cv=5)

clf.fit(X_train_norm, y_train)
scores = clf.cv_results_['mean_test_score']

print("After Scaling")
for score, solver in zip(scores, solver_list):
    print(f"  {solver} {score:.3f}" )
```

```python
log_clf_lbfgs_norm=LogisticRegression(solver='lbfgs')
log_clf_lbfgs_norm.fit(X_train_norm,y_train)
print_score(log_clf_lbfgs_norm,X_train_norm, y_train, X_test_norm, y_test, train=True)
print_score(log_clf_lbfgs_norm,X_train_norm, y_train, X_test_norm, y_test, train=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
y_predict_norm=log_clf_lbfgs_norm.predict(X_test_norm)
acc3=accuracy_score(y_test, y_predict_norm) * 100
conf_matrix = confusion_matrix(y_test, y_predict_norm)
plt.figure(figsize=(4, 4))
sns.heatmap(conf_matrix,  annot=True,xticklabels=['Neg','pos'], yticklabels=['Neg','Pos'],fmt="d")
plt.title("Confusion matrix_With Normalization\nSolver:lbfgs, Accuracy=%.2f"%acc3)
plt.ylabel('True class')
plt.xlabel('Predicted class')
plt.show()
```

```python
log_clf_lbfgs_norm.coef_
```

```python
log_clf_liblinear_norm=LogisticRegression(solver='liblinear')
log_clf_liblinear_norm.fit(X_train_norm,y_train)
print_score(log_clf_liblinear_norm,X_train_norm, y_train, X_test_norm, y_test, train=True)
print_score(log_clf_liblinear_norm,X_train_norm, y_train, X_test_norm, y_test, train=False)
```
